```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

messages = [
"Congratulations you won a lottery",
"Win cash now click here",
"You have been selected for a free gift",
"Hello how are you",
"Are you coming to college",
"Let's meet tomorrow",
"Free entry in contest",
"Call this number to claim your reward",
"Can we meet today",
"Please send the report",
"offer"
]

labels = [1,1,1,0,0,0,1,1,0,0,1]

cv = CountVectorizer()
X = cv.fit_transform(messages)

X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.2, random_state=42
)

model = MultinomialNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
plt.figure()
plt.bar(["Accuracy"], [accuracy])
plt.title("Spam Detection Model Accuracy")
plt.ylabel("Accuracy Score")
plt.ylim(0, 1)
plt.show()
user_input = input("Enter your message: ")
input_vec = cv.transform([user_input])
prediction = model.predict(input_vec)
print("\nPrediction:", "Spam" if prediction[0] == 1 else "Not Spam")
```
